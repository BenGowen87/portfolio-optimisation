```python
import yfinance as yf #library used to install stock market data from Yahoo finance
import matplotlib.pyplot as plt #used for creating graphs

#downloads the companys stock data from 2019 to 2024
aapl = yf.download("AAPL", start="2019-01-01", end="2024-01-01")
msft = yf.download("MSFT", start="2019-01-01", end="2024-01-01")
nvda = yf.download("NVDA", start="2019-01-01", end="2024-01-01")
spy = yf.download("SPY", start="2019-01-01", end="2024-01-01")

#creates the size of the graph
plt.figure(figsize=(12,6))

#plots the date on the X-axis and the corresponding price in the Close field on the Y-axis
#the dataframe is sorted by dates so you can use .index
plt.plot(aapl.index, aapl["Close"], label="Apple")
plt.plot(msft.index, msft["Close"], label="Microsoft")
plt.plot(nvda.index, nvda["Close"], label="Nvidia")
plt.plot(spy.index, spy["Close"], label="S&P 500")

#creates the titles of the graph and the legend
plt.title("Stock Prices")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend() #shows which graph corresponds to what company

plt.show()
```
