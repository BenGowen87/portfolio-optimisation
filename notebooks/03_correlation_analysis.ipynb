```python
from src.data_utils import get_returns_data

#get the returns data frame
returns_df = get_returns_data()

#prints the first few dates
print(returns_df.head())
```

```python
correlation_matrix = returns_df.corr()

print(correlation_matrix)
```

This is the correlation matrix which represents the correlation between each pair of stocks. We notice that this matrix is symetric as the correlation between Apple and Microsoft is the same as the correlation between Microsoft and Apple. Correlation is a result between -1 and 1 which measures the strength and direction of the relationship between 2 variables. A correlation coefficient of 1 means theres a perfect linear relationship between the 2 variables (if one increases the other increases by a predicatable amount). A correlation coefficient of 0 means there is no linear relationship between the 2 variables.

From the correlation matrix you can see that the correlation between the S&P500 and the 3 companies is a strong positive linear relationship as the S&P500 reflects the general market performance. Additionally the correlation between Apple and Microsoft is a strong positive relationship as they are both large technology companies that are likely influenced by similar market conditions.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# creating a heat map of the correlation matrix
plt.figure(figsize=(8,6))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Matrix")

plt.show()
```

Correlation is important because investing in stocks with imperfect correlation can reduce portfolio risk. If all assets in a portfolio behave similarly, a decrease in one asset’s returns is likely to coincide with decreases in the others.
